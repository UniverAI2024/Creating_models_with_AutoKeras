# Тюнеры

В библиотеке AutoKeras есть несколько тюнеров, которые могут быть использованы для автоматического подбора гиперпараметров моделей машинного обучения.

**Random (random):**
Случайный поиск является простейшим подходом к оптимизации гиперпараметров. Он просто выбирает случайные комбинации гиперпараметров из заданного распределения и обучает модель для каждой комбинации. Однако его главным недостатком является то, что он может быть менее эффективным, чем другие методы, особенно когда пространство гиперпараметров велико.

**Bayesian (bayesian):**
Байесовская оптимизация является более сложным и интеллектуальным подходом к оптимизации гиперпараметров. Он использует вероятностную модель для оценки качества различных комбинаций гиперпараметров. Вместо случайного выбора комбинаций гиперпараметров, байесовский тюнер выбирает их на основе предыдущих результатов и текущего состояния вероятностной модели.

Основная идея Bayesian тюнера заключается в том, что он создает модель, которая прогнозирует, как изменение гиперпараметров влияет на функцию ошибки нейросети. Для этого он использует гауссовский процесс, который представляет собой модель, описывающую распределение функции ошибки в зависимости от значений гиперпараметров. Гауссовский процесс может быть использован для оценки потенциальных значений функции ошибки на новых наборах гиперпараметров, не обучая нейросеть на каждой итерации.

Алгоритм работы Bayesian тюнера можно описать следующим образом:

- Определить пространство поиска гиперпараметров и задать функцию ошибки, которую мы хотим минимизировать.

- Инициализировать гауссовский процесс, используя начальный набор гиперпараметров и соответствующие значения функции ошибки.

- Повторять следующие шаги до тех пор, пока не будет достигнуто максимальное количество итераций:

- Используя гауссовский процесс, выбрать следующий набор гиперпараметров, который предположительно даст наименьшее значение функции ошибки.

- Обучить нейросеть на выбранном наборе гиперпараметров и вычислить значение функции ошибки на тестовом наборе данных.

- Обновить модель гауссовского процесса, используя новый набор гиперпараметров и соответствующее значение функции ошибки.

- Выбрать набор гиперпараметров с наименьшим значением функции ошибки на тестовом наборе данных.

Байесовская оптимизация может быть более эффективной, чем случайный поиск, особенно в случае малого количества обучающих примеров или небольшого пространства гиперпараметров. Однако, байесовский тюнер может быть более медленным, так как его обновление модели может занять некоторое время.

**Hyperband (hyperband):**
Hyperband является адаптивным алгоритмом оптимизации гиперпараметров, основанным на идее адаптивного выделения ресурсов для различных комбинаций гиперпараметров. Вместо одинакового выделения ресурсов для каждой комбинации, Hyperband адаптируется на основе предыдущих результатов и выделяет больше ресурсов для комбинаций с большим потенциалом. Hyperband комбинирует две стратегии: диверсификация (начиная с множества случайных комбинаций и малого количества эпох) и интенсификация (продолжая обучение лучших комбинаций на большем количестве эпох). Это делает Hyperband более эффективным и адаптивным, чем случайный поиск или байесовская оптимизация. Однако, он также может быть более сложным и требовать больше вычислительных ресурсов, особенно на начальных этапах.

Основная идея Hyperband заключается в том, что он запускает множество случайных поисков на разных уровнях ресурсов (число эпох, используемых для обучения), и эффективно прерывает худшие поиски для освобождения ресурсов, которые могут быть использованы для более перспективных поисков.

Алгоритм Hyperband состоит из нескольких шагов:

- Инициализация набора случайных конфигураций гиперпараметров и определение максимального количества ресурсов, которые будут выделены на каждый поиск.

- Запуск нескольких случайных поисков на разных уровнях ресурсов, начиная с наименьшего количества ресурсов. На каждом уровне ресурсов, все модели обучаются на данном уровне ресурсов в течение нескольких эпох, и затем наихудшая половина моделей отбрасывается.

- Повторение шага 2 на следующем уровне ресурсов с меньшим количеством моделей, но большим количеством эпох, пока не будет достигнуто максимальное количество ресурсов или не останется только одна модель.

- Оценка точности лучшей модели на тестовом наборе данных, используя найденные оптимальные гиперпараметры.

Один из главных преимуществ Hyperband заключается в том, что он позволяет быстро находить оптимальные гиперпараметры на больших наборах данных, используя меньше вычислительных ресурсов, чем другие алгоритмы оптимизации гиперпараметров

**Greedy (greedy):**
Жадный тюнер является методом последовательного выбора гиперпараметров. Он начинает с настройки базовых гиперпараметров, а затем последовательно настраивает каждый гиперпараметр, выбирая тот, который приводит к наибольшему улучшению результатов. Преимущества жадного подхода включают простоту и возможность быстрого улучшения результатов. Однако этот подход может привести к локальному оптимуму, так как он не учитывает возможные взаимодействия между гиперпараметрами.

В случае жадного настройщика, он начинает с некоторой начальной комбинации гиперпараметров и последовательно изменяет каждый гиперпараметр на заданный шаг изменения, сохраняя лучшую комбинацию гиперпараметров, которую он находит на каждом шаге. Он выбирает локально оптимальное решение на каждом шаге, не учитывая будущие последствия, таким образом, жадный настройщик может застрять в локальном оптимуме и не сможет достичь глобального оптимума.

Каждый тюнер имеет свои уникальные преимущества и недостатки, и лучший тюнер для вашей задачи может зависеть от многих факторов, таких как размер выборки данных, количество гиперпараметров, которые нужно подобрать, и доступные ресурсы. Общий подход состоит в том, чтобы попробовать несколько тюнеров и выбрать тот, который дает наилучшие результаты в вашей конкретной задаче.

### Запускаем классификатор

Для классификации изображений в Autokeras используется ImageClassifier, где при помощи параметров нами устанавливаются необходимые данные, касающиеся самого датасета, обучения и моделей.

**max_trials**: максимальное количество раз, которое AutoKeras будет искать модели (по умолчанию 100).

**directory**: путь к папке, где AutoKeras будет сохранять результаты (по умолчанию 'auto-keras').

**objective**: целевая метрика, которую AutoKeras будет оптимизировать (по умолчанию 'val_loss').

**seed**: начальное значение для генератора случайных чисел (по умолчанию None). В AutoKeras ImageClassifier параметр seed используется при инициализации некоторых внутренних компонентов, которые могут влиять на обучение модели, таких как способ инициализации весов нейросети или случайное разбиение набора данных на обучающую и тестовую выборки.

В общем случае, если параметр seed не установлен, то каждый запуск программы будет давать разные результаты. Если требуется повторяемость результатов, то следует установить один и тот же seed при каждом запуске программы.

**max_model_size**: максимальный размер модели.

**num_classes**: количество классов в задаче классификации (может определяться автоматически)

**multi_label**: флаг, указывающий, является ли задача классификации многоклассовой (по умолчанию False). Мульти-лейбл классификация - это задача классификации, где объект может быть отнесен к нескольким классам одновременно. Например, в задаче классификации изображений, где на изображении может быть несколько объектов, каждый объект может быть отнесен к своему классу. В этом случае мульти-лейбл классификация будет более подходящим выбором.

**loss**: функция потерь для обучения модели (по умолчанию 'categorical_crossentropy' для многоклассовой классификации и 'binary_crossentropy' для бинарной классификации).

**metrics**: список метрик для оценки качества модели (по умолчанию ['accuracy']).

**project_name**: имя проекта для сохранения моделей (по умолчанию 'image_classifier').

**tuner**: Оптимизатор для подбора архитектуры модели. Можно выбрать 'greedy', 'bayesian', 'hyperband' или 'random'. По умолчанию используется тюнер для конкретных задач, который сначала оценивает наиболее часто используемые модели для задачи, прежде чем исследовать другие модели.

**overwrite**: по умолчанию "False",  если True, удаляет предыдущие результаты и начинает заново
   
**kwargs**:  Любые аргументы, поддерживаемые AutoModel.

Вы можете использовать эти параметры для настройки AutoKeras в соответствии с вашими потребностями.

В метод **fit** можно передать следующие аргументы:

**x**: массив или массивы массивов входных данных в формате numpy.ndarray либо tensorflow.Dataset.

**y**: массив целевых меток. Должен иметь такую же длину, как и x.

**epochs**: количество эпох для обучения модели.

**callbacks**: список объектов keras.callbacks.Callback для вызова во время обучения.

**validation_split**: доля данных, которые будут использоваться для проверки модели в процессе обучения. Должно быть в диапазоне от 0 до 1.

**validation_data**: кортеж (x_val, y_val) или массив (x_val, y_val) с проверочными данными.

****kwargs**: дополнительные аргументы для метода fit в TensorFlow Keras.

Метод **fit** выполняет тренировку модели, используя заданные входные данные и метки. Он также может использовать данные проверки для оценки производительности модели в процессе обучения.

В данном эксперименте мы запускаем ImageClassifier с минимальным количеством параметров. Количество классов, соответствующая функция ошибки и метрика определяется автоматически. Если не указывать тюнер, то по умолчанию будет использован тюнер для конкретных задач, который запускает обучение на наиболее часто используемых модели для подобных задач.

```python
# устанавливаем Autokeras
!pip install autokeras
```

```python
import autokeras as ak

# Инициализация классификатора AutoKeras
clf = ak.ImageClassifier(
    max_trials=2,
    objective='val_accuracy',
    directory='pavement_crack',
    overwrite=True)

clf.fit(X_train, y_train, epochs=8, validation_data=(X_val, y_val))
```

### Получаем архитектуру модели и используемые в ней параметры

После обучения ak.ImageClassifier, вы можете получить лучшую модель, а затем изучить её параметры и структуру. Вот как это можно сделать:


```python
best_model = clf.export_model()
# Получите архитектуру модели в виде JSON строки
model_json = best_model.to_json()

print("Model architecture:")
print(model_json)
```

```python
best_model.summary()
```

```python
# для более детального изучение параметров модели,можно итерировать по слоям модели:
print("\nLayer details:")
for layer in best_model.layers:
    print(f"Layer name: {layer.name}")
    print(f"Layer type: {type(layer)}")
    config = layer.get_config()
    print("Layer configuration:")
    for key, value in config.items():
        print(f"  {key}: {value}")

optimizer = best_model.optimizer
learning_rate = optimizer.lr
print(f"\nLearning rate: {learning_rate}")
```

*x*: массив или массивы массивов входных данных для предсказания.

verbose=1: выводить индикатор прогресса на каждую эпоху обучения.

verbose=2: выводить информацию о каждой эпохе обучения.

**evaluate:**

*x*: массив или массивы массивов входных данных для оценки.

*y*: массив целевых меток. Должен иметь такую же длину, как и x.

*batch_size*: количество образцов, передаваемых в одной итерации для оценки. Если не указано, значение batch_size по умолчанию равно 32.

*verbose*: управляет количеством выводимой информации на экран во время выполнения метода.

AutoModel в AutoKeras является классом для автоматического создания и обучения моделей глубокого обучения. Этот класс используется для определения пользовательской архитектуры модели и настройки гиперпараметров.

В отличие от предустановленных классов, таких как ImageClassifier, TextClassifier и других, AutoModel предоставляет больше гибкости и контроля над архитектурой модели. Вы можете создать кастомную архитектуру , определить свои входы и выходы, а затем передать их в экземпляр AutoModel.

### Эксперимент №1: Automodel с нормализацией и сверточными блоками

```python
import autokeras as ak

# Создаем входной узел для изображений
input_node = ak.ImageInput()

# Добавляем слой нормализации к входному узлу
output_node = ak.Normalization()(input_node)

# Добавляем сверточный блок с двумя слоями и 128 фильтрами
output_node = ak.ConvBlock(num_layers=2, filters=128)(output_node)

# Добавляем регрессионный слой
output_node = ak.RegressionHead()(output_node)

# Создаем экземпляр AutoModel с нашими кастомными входами и выходами
clf = ak.AutoModel(
    inputs=input_node,
    outputs=output_node,
    overwrite=True,  # Перезаписать предыдущие результаты при повторном запуске
    max_trials=3    # Максимальное количество попыток для каждой модели
)

# Обучаем модель на тренировочных данных и проверяем на валидационных данных
clf.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val))
```

```python
#Получаем лучшую модель:
best_model = clf.export_model()
# Получите архитектуру модели в виде JSON строки
model_json = best_model.to_json()

print("Model architecture:")
print(model_json)
```

```python
# Получаем структуру лучшей модели
best_model.summary()
```

Слой SeparableConv2D отличается от Conv2D тем, что он использует два этапа свертки вместо одного, что позволяет снизить количество параметров и ускорить вычисления.

Первый этап свертки - Depthwise Convolution - применяется к каждому входному каналу отдельно. Это означает, что каждый входной канал свертывается с собственным фильтром. В результате получается такое же количество выходных каналов, что и входных. Таким образом, количество параметров для этого этапа свертки сравнительно невелико.

Второй этап свертки - Pointwise Convolution - применяется к выходу Depthwise Convolution. В этом этапе используется свертка 1x1, которая позволяет соединять различные каналы входных данных и создавать новые выходные каналы. Этот этап свертки позволяет сгенерировать любое количество выходных каналов.

AutoModel в AutoKeras позволяет указать структуру модели, определяя входные и выходные узлы, а также основные слои и блоки между ними. Затем AutoKeras автоматически подбирает оптимальные гиперпараметры для указанных слоев и блоков в процессе поиска модели.

Например, в предыдущем коде мы указали использовать сверточный блок с определенным количеством слоев и фильтров. AutoKeras затем самостоятельно определит оптимальные гиперпараметры, такие как размер ядра, тип активации и использование нормализации пакета внутри каждого сверточного блока.

### Блоки для создания пользовательских моделей

AutoKeras предоставляет множество различных блоков для создания пользовательских моделей. Вот несколько из них:

**ak.Normalization()**: это блок предварительной обработки в AutoKeras, который выполняет нормализацию данных. Нормализация - это процесс масштабирования данных таким образом, чтобы они имели среднее значение 0 и стандартное отклонение 1. Это полезно для обучения глубоких нейронных сетей, поскольку масштабирование данных облегчает обучение модели и улучшает сходимость.

**ak.ConvBlock()**: это блок в AutoKeras, который создает слой свертки с настраиваемыми параметрами. Он позволяет создавать сверточные слои с различными параметрами, такими как число фильтров, размер ядра, функция активации и другие.

**ak.ResNetBlock**: Реализует блок ResNet с опциональным количеством слоев и определенными значениями фильтров.

**ak.XceptionBlock**: Реализует блок Xception, который является основой архитектуры Xception, с использованием глубоких сверток и отдельных сверток.

**ak.DenseBlock**: Реализует блок плотных (полносвязных) слоев с определенным количеством слоев и нейронов.

**ak.SpatialReduction**: Уменьшает размерность пространственных данных, таких как изображения или текст, с использованием GlobalAveragePooling или GlobalMaxPooling.

**ak.Merge**: Объединяет два или более входных узлов в один выходной узел с использованием различных методов слияния, таких как конкатенация, сложение.

## Параметры TextClassifier

**ak.TextClassifier** - это класс библиотеки autokeras, который используется для автоматического обучения моделей классификации текста.

Некоторые из основных параметров, которые можно настроить в ak.TextClassifier, включают:

**max_trials**: максимальное число различных конфигураций модели, которые будут рассмотрены при поиске оптимальной модели

**objective**: метрика, используемая для оценки качества модели и выбора наилучшей конфигурации. В autokeras.TextClassifier значение параметра objective по умолчанию устанавливается в val_loss. Он используется для выбора модели с минимальным значением функции потерь на валидационной выборке данных.

**multi_label**: логическое значение, если multi_label=True, то это означает, что каждый текст может иметь несколько меток классов. Например, в задаче классификации текстов по темам, один документ может относиться к нескольким темам одновременно. Если multi_label=False (по умолчанию), то каждый текст должен иметь только одну метку класса.

**num_classes** - это параметр, который указывает количество классов, на которые необходимо классифицировать данные. Если не указывать - количество классов определится автоматически

**loss** - это функция потерь (loss function), используемая при обучении модели классификации текста в autokeras.TextClassifier.

Функция потерь измеряет, насколько точно модель предсказывает метки классов на обучающем наборе данных и используется для оптимизации параметров модели в процессе обучения.

В autokeras.TextClassifier значение параметра loss по умолчанию устанавливается в categorical_crossentropy для задачи многоклассовой классификации и в binary_crossentropy для задачи мультилейбл классификации.

**project_name** - это параметр в классе autokeras.TextClassifier, который определяет название проекта, используемого для сохранения модели.

Параметр project_name позволяет сохранять автоматически созданные модели классификации текста с именами, которые вы можете легко идентифицировать. Это может быть полезно, когда вы обучаете несколько моделей и хотите сохранять их в разных проектах.

**directory** - это параметр , который определяет путь к директории, в которой будут сохранены все результаты, связанные с обучением и использованием модели.

**tuner** - это параметр в классе autokeras.TextClassifier, который определяет тип оптимизатора, используемого для настройки архитектуры и параметров модели. autokeras поддерживает несколько различных типов оптимизаторов, в том числе BayesianOptimization, Hyperband, random и Greedy. Если его не указать, он использует настройщик для конкретной задачи, который сначала оценивает наиболее часто используемые модели для задачи, прежде чем исследовать другие модели.

**overwrite** - это параметр, который определяет, будет ли существующая модель перезаписана при повторном обучении с тем же именем проекта. По умолчанию False.

Если overwrite=True, существующая модель будет перезаписана при повторном обучении. Это может быть полезно, если вы хотите заменить существующую модель более новой или обучить модель заново с теми же параметрами.

**seed** - это параметр в классе autokeras.TextClassifier, который определяет начальное значение генератора случайных чисел, используемого при обучении модели.

Параметр seed позволяет воспроизводить результаты обучения модели при каждом запуске, что может быть полезно для тестирования и отладки моделей.

В AutoKeras, есть три типа блоков для работы с текстами: 'vanilla', 'transformer' и 'ngram'.

**Vanilla** - это самый простой тип блока. Он состоит из одного слоя эмбеддинга, за которым следуют несколько сверточных слоев, которые сжимают и агрегируют признаки в последовательности. Этот тип блока хорошо работает для задач классификации текста с короткими последовательностями, такими как твиты или заголовки новостей.

**Transformer** - это тип блока, основанный на архитектуре трансформера. Он состоит из нескольких слоев энкодера, которые используют механизм внимания, чтобы получить контекстную информацию из всей последовательности. Этот тип блока лучше всего подходит для задач генерации текста или машинного перевода.

**Ngram** - это тип блока, который использует n-граммы для извлечения признаков из последовательности. Он состоит из слоя эмбеддинга, за которым следует несколько слоев конволюционной нейронной сети. Этот тип блока хорошо работает для задач классификации текста с длинными последовательностями, таких как отзывы на фильмы или описания товаров.

Выбор типа блока зависит от задачи, которую вы хотите решить, и от особенностей входных данных. Если вы работаете с короткими последовательностями, то Vanilla может быть хорошим выбором. Если вы работаете с длинными последовательностями или задачами генерации текста, то Transformer или Ngram могут быть лучшими вариантами.

## Параметры StructuredDataRegressor

**Autokeras.StructuredDataRegressor** - это класс автоматической машинной обучения в AutoKeras для задачи регрессии на структурированных данных. Вот некоторые из основных параметров, которые можно использовать при создании объекта autokeras.StructuredDataRegressor:

**column_names**: список имен столбцов входных данных. По умолчанию равен None (будет получен из заголовка CSV-файла или pandas.DataFrame).

**column_types**: список типов столбцов входных данных. По умолчанию равен None.

# Autokeras для работы с временными рядами

Для демонстрации работы Autokeras с временными рядами используем датасет стоимости акций «Лукойла» за несколько лет.

**ak.TimeseriesForecaster** - это класс библиотеки autokeras, который используется для автоматического обучения моделей прогнозирования временных рядов.

Этот класс предназначен для прогнозирования временных рядов с помощью методов машинного обучения, в том числе рекуррентных нейронных сетей, сверточных нейронных сетей и других алгоритмов. ak.TimeseriesForecaster использует автоматический подбор гиперпараметров и оптимизацию модели, чтобы найти наилучшую архитектуру модели для задачи прогнозирования временных рядов.

Некоторые из основных параметров, которые можно настроить в ak.TimeseriesForecaster, включают:

**lookback**: число временных шагов, используемых для прогнозирования будущих значений временного ряда

**max_trials**: максимальное число различных конфигураций модели, которые будут рассмотрены при поиске оптимальной модели

**objective:** метрика, используемая для оценки качества модели и выбора наилучшей конфигурации

Параметры **predict_from** и **predict_until** используются для определения диапазона индексов временных шагов, для которых необходимо сделать прогноз с помощью модели. Например, predict_from=1 и predict_until=10 указывают на то, что модель должна сделать прогноз для временных шагов от индекса 1 до индекса 10 во временном ряду.

Класс также предоставляет ряд методов для обучения и использования модели, включая fit() и predict().

# Multi-Modal модель

Multi-Modal в Autokeras - это возможность обучать модели на нескольких входах данных, которые могут иметь различную структуру, например, текст, изображения, числовые признаки и т.д. Эта функциональность называется "мультимодальностью", и она позволяет создавать более сложные модели, которые могут использовать несколько источников информации для более точных прогнозов.

Multi-Modal в Autokeras позволяет комбинировать различные типы данных в единую модель с помощью графа вычислений, который задает структуру модели. Каждый входной тип данных может иметь свой отдельный граф вычислений. Эти графы могут быть объединены, чтобы обучить единую модель, используя различные комбинации данных и графов.

Например, можно создать модель, которая принимает на вход табличные и текстовые данные, и использует их для предсказания цены. Можно создать два графа вычислений для каждого типа входных данных, которые затем будут объединены в единую модель.

```python
#Во всех функция дальше
#d - описание квартиры, одна строка из data1

#Вычисляем количество комнат
#maxRoomCount - максимальное число комнат в квартире
def getRoomsCount(d, maxRoomCount):
  roomsCountStr = d[0] #Получаем строку с числом комнат

  roomsCount = 0
  try:
    roomsCount = int(roomsCountStr) #Пробуем превратить строку в число
    if (roomsCount > maxRoomCount):
      roomsCount = maxRoomCount #Если число комнат больше максимального, то присваиваем максимальное
  except: #Если не получается превратить строку в число
    if (roomsCountStr == roomsCountStr): #Проверяем строку на nan (сравнение с самим собой)
      if ("Ст" in roomsCountStr): #Еcть строка = "Ст", значит это Студия
        roomsCount = maxRoomCount + 1

  return roomsCount

#Превращаем число комнат в категорию
def getRoomsCountCategory(d, maxRoomCount):
  roomsCount = getRoomsCount(d, maxRoomCount) #Получаем число комнат
  roomsCount = utils.to_categorical(roomsCount, maxRoomCount+2) #Превращаем в категорию
  #maxRoomCount+2 потому что 0 зарезервирован на неопознаное число комнат, а maxRoomCount+1 на "Студию"
  return roomsCount

#Получаем индекс станции метро
#allMetroNames - все уникальные названия метро в базе
def getMetro(d, allMetroNames):
  metroStr = d[1] #Получаем строку метро
  metro = 0

  if (metroStr in allMetroNames): #Если находим метро во всех названиях
    metro = allMetroNames.index(metroStr)+1 #Присваиваем индекс
    #+1 так как 0 зарезервирован на неопознанное метро

  return metro

#Получаем тип метро
#0 - внутри кольца
#1 - кольцо
#2 - 1-3 станции от конца
#3 - 4-8 станций от кольца
#4 - больше 8 станций от кольца
def getMetroType(d):
  metroTypeStr = d[1] #Получаем строку метро
  metroTypeClasses = 5 #Число классов метро
  metroType = metroTypeClasses - 1 #Изначально считаем последний класс

  #Метро внутри кольца
  metroNamesInsideCircle = ["Площадь Революции", "Арбатская", "Смоленская", "Красные Ворота", "Чистые пруды", "Лубянка", "Охотный Ряд", "Библиотека имени Ленина", "Кропоткинская", "Сухаревская", "Тургеневская", "Китай-город", "Третьяковская", "Трубная", "Сретенский бульвар", "Цветной бульвар", "Чеховская", "Боровицкая", "Полянка", "Маяковская", "Тверская", "Театральная", "Новокузнецкая", "Пушкинская", "Кузнецкий Мост", "Китай-город", "Александровский сад"]
  #Метро на кольце
  metroNamesCircle = ["Киевская", "Парк Культуры", "Октябрьская", "Добрынинская", "Павелецкая", "Таганская", "Курская", "Комсомольская", "Проспект Мира", "Новослободская", "Белорусская", "Краснопресненская"]
  #Метро 1-3 станции от кольца
  metroNames13FromCircle = ["Бауманская", "Электрозаводская", "Семёновская", "Площадь Ильича", "Авиамоторная", "Шоссе Энтузиастов", "Римская", "Крестьянская Застава", "Дубровка", "Пролетарская", "Волгоградский проспект", "Текстильщики", "Автозаводская", "Технопарк", "Коломенская", "Тульская", "Нагатинская", "Нагорная", "Шаболовская", "Ленинский проспект", "Академическая", "Фрунзенская", "Спортивная", "Воробьёвы горы", "Студенческая", "Кутузовская", "Фили", "Парк Победы", "Выставочная", "Международная", "Улица 1905 года", "Беговая", "Полежаевская", "Динамо", "Аэропорт", "Сокол", "Деловой центр", "Шелепиха", "Хорошёвская", "ЦСКА", "Петровский парк", "Савёловская", "Дмитровская", "Тимирязевская", "Достоевская", "Марьина Роща", "Бутырская", "Фонвизинская", "Рижская", "Алексеевская", "ВДНХ", "Красносельская", "Сокольники", "Преображенская площадь"]
  #Метро 4-8 станций от кольа
  metroNames48FromCircle = ["Партизанская", "Измайловская", "Первомайская", "Щёлковская", "Новокосино", "Новогиреево", "Перово", "Кузьминки", "Рязанский проспект", "Выхино", "Лермонтовский проспект", "Жулебино", "Партизанская", "Измайловская", "Первомайская", "Щёлковская", "Новокосино", "Новогиреево", "Перово", "Кузьминки", "Рязанский проспект", "Выхино", "Лермонтовский проспект", "Жулебино", "Улица Дмитриевского", "Кожуховская", "Печатники", "Волжская", "Люблино", "Братиславская", "Коломенская", "Каширская", "Кантемировская", "Царицыно", "Орехово", "Севастопольская", "Чертановская", "Южная", "Пражская", "Варшавская", "Профсоюзная", "Новые Черёмушки", "Калужская", "Беляево", "Коньково", "Университет", "Багратионовская", "Филёвский парк", "Пионерская", "Кунцевская", "Молодёжная", "Октябрьское Поле", "Щукинская", "Спартак", "Тушинская", "Сходненская", "Войковская", "Водный стадион", "Речной вокзал", "Беломорская", "Ховрино", "Петровско-Разумовская", "Владыкино", "Отрадное", "Бибирево", "Алтуфьево", "Фонвизинская", "Окружная", "Верхние Лихоборы", "Селигерская", "ВДНХ", "Ботанический сад", "Свиблово", "Бабушкинская", "Медведково", "Преображенская площадь", "Черкизовская", "Бульвар Рокоссовского"]

  #Проверяем, в какую категорию попадает наша станция
  if (metroTypeStr in metroNamesInsideCircle):
    metroType = 0
  if (metroTypeStr in metroNamesCircle):
    metroType = 1
  if (metroTypeStr in metroNames13FromCircle):
    metroType = 2
  if (metroTypeStr in metroNames48FromCircle):
    metroType = 3

  #Превращаем результат в категорию
  metroType = utils.to_categorical(metroType, metroTypeClasses)
  return metroType


#Вычисляем растояние до метро
def getMetroDistance(d):
  metroDistanceStr = d[2] #Получаем строку

  metroDistance = 0 #Расстояние до метро
  metroDistanceType = 0 #Тип расстояния - пешком или на транспорте

  #ЕСли строка не равна nan
  if (metroDistanceStr == metroDistanceStr):
    if (len(metroDistanceStr) > 0):
      #Определяем тип расстояния
      if (metroDistanceStr[-1] == "п"):
        metroDistanceType = 1 #Пешком
      elif (metroDistanceStr[-1] == "т"):
        metroDistanceType = 2 #На транспорте

      #Выбрасываем последний символ, чтобы осталось только число
      metroDistanceStr = metroDistanceStr[:-1]
      try:
        #Разделяем дистанции на категории
        metroDistance = int(metroDistanceStr)
        if (metroDistance < 3):
          metroDistance = 1
        elif (metroDistance < 6):
          metroDistance = 2
        elif (metroDistance < 10):
          metroDistance = 3
        elif (metroDistance < 15):
          metroDistance = 4
        elif (metroDistance < 20):
          metroDistance = 5
        else:
          metroDistance = 6
      except: #Если в строке не число, то категория 0
        metroDistance = 0

  #Число классов дистанции
  metroDistanceClasses = 7

  #У нас 7 категорий дистанции по расстоянию
  #И 3 типа дистанции - неопознанный, пешком и транспортом
  #Мы создадим вектор длины 3*7 = 21
  #Будем преобразовывать индекс расстояния 0-6 в 0-20
  #Для типа "Пешком" - ничего не меняем
  if (metroDistanceType == 2):
    metroDistance += metroDistanceClasses #Для типа "Транспортом" добавляем 7
  if (metroDistanceType == 0):
    metroDistance += 2*metroDistanceClasses #Для неопознанного типа добавляем 14

  #Превращаем в категории
  metroDistance = utils.to_categorical(metroDistance, 3*metroDistanceClasses)
  return metroDistance

#Получаем 4 данных
#- этаж квартиры
#- этажность дома
#- индикатор, что последний этаж
#- тип дома
def getHouseTypeAndFloor(d):
  try:
    houseStr = d[3] #Получаем строку типа дома и этажей
  except:
    houseStr = ""

  houseType = 0 #Тип дома
  floor = 0 #Этаж квартиры
  floors = 0 #Этажность дома
  isLastFloor = 0 #Индикатор последнего этажа

  #Проверяем строку на nan
  if (houseStr == houseStr):
    if (len(houseStr) > 1):

      try:
        slashIndex = houseStr.index("/") #Ищем разделитель /
      except:
        print(houseStr)

      try:
        spaceIndex = houseStr.index(" ") #Ищем разделитель " "
      except:
        print(houseStr)

      #Вытаскиваем строки
      floorStr = houseStr[:slashIndex] #Строка этажа
      floorsStr = houseStr[slashIndex+1:spaceIndex] #Строка этажнгости дома
      houseTypeStr = houseStr[spaceIndex+1:] #Строка типа дома

      #Выбираем категорию этажа
      try:
        floor = int(floorStr) #Превращаем строку в число
        floorSave = floor
        if (floorSave < 5):
          floor = 2
        if (floorSave < 10):
          floor = 3
        if (floorSave < 20):
          floor = 4
        if (floorSave >= 20):
          floor = 5
        if (floorSave == 1): #Первый этаж выделяем в отдельную категорию
          floor = 1

        if (int(floorStr) == int(floorsStr)): #Если этаж последний, включаем индикатор последнего этажа
          isLastFloor = 1
      except:
        floor = 0 #Если строка не парсится в число, то категория этажа = 0 (отдельная)

      #Выбираем категорию этажности дома
      try:
        floors = int(floorsStr) #Превращаем строку в число
        floorsSave = floors
        if (floorsSave < 5):
          floors = 1
        if (floorsSave < 10):
          floors = 2
        if (floorsSave < 20):
          floors = 3
        if (floorsSave >= 20):
          floors = 4
      except:
        floors = 0 #Если строка не парсится в число, то категория этажности = 0 (отдельная)

      #Определяем категорию типа дома
      if (len(houseTypeStr) > 0):
        if ("М" in houseTypeStr):
          houseType = 1
        if ("К" in houseTypeStr):
          houseType = 2
        if ("П" in houseTypeStr):
          houseType = 3
        if ("Б" in houseTypeStr):
          houseType = 4
        if ("?" in houseTypeStr):
          houseType = 5
        if ("-" in houseTypeStr):
          houseType = 6

    #Превращаем все категории в one hot encoding
    floor = utils.to_categorical(floor, 6)
    floors = utils.to_categorical(floors, 5)
    houseType = utils.to_categorical(houseType, 7)


  return floor, floors, isLastFloor, houseType


#Вычисляем тип балкона
def getBalcony(d):
  balconyStr = d[4] #Полуаем строку
  #Выписываем все варианты балконов в базе
  balconyVariants = ['Л', 'Б', '2Б', '-', '2Б2Л', 'БЛ', '3Б', '2Л', 'Эрк', 'Б2Л', 'ЭркЛ', '3Л', '4Л', '*Л', '*Б']
  #Проверяем на nan
  if (balconyStr == balconyStr):
    balcony = balconyVariants.index(balconyStr)+1 #Находим индекс строки балкона во всех строках
  else:
    balcony = 0 #Индекс 0 выделяем на строку nan

  #Превращаем в one hot encoding
  balcony = utils.to_categorical(balcony, 16)

  return balcony

#Определяем тип санузла
def getWC(d):
  wcStr = d[5] #Получаем строку
  #Выписываем все варианты санузлов в базе
  wcVariants = ['2', 'Р', 'С', '-', '2С', '+', '4Р', '2Р', '3С', '4С', '4', '3', '3Р']
  #Проверяем на nan
  if (wcStr == wcStr):
    wc = wcVariants.index(wcStr)+1 #Находим индекс строки санузла во всех строках
  else:
    wc = 0 #Индекс 0 выделяем на строку nan

  #Превращаем в one hot encoding
  wc = utils.to_categorical(wc, 14)

  return wc

#Определяем площадь
def getArea(d):
  areaStr = d[6] #Полуачем строку площади

  if ("/" in areaStr):
    slashIndex = areaStr.index("/") #Находим разделитель /
    try:
      area = float(areaStr[:slashIndex]) #Берём число до разделителя и превращаем в число
    except:
      area = 0 #Если не получается, возвращаем 0
  else:
    area = 0 #Или если нет разделителя, возвращаем 0

  return area

#Получаем цену
def getCost(d):
  costStr = d[7] #Загружаем строку

  try:
    cost = float(costStr) #Пробуем превратить в число
  except:
    cost = 0 #Если не получается, возвращаем 0

  return cost

#Получаем комментарий
def getComment(d):
  commentStr = d[-1] #Возвращаем данные из последней колонки

  return commentStr

#Объединяем все числовые параметры вместе
def getAllParameters(d, allMetroNames):
  #Загружаем все данные по отдельности
  roomsCountType = getRoomsCountCategory(d, 30)
  metro = getMetro(d, allMetroNames)
  metroType = getMetroType(d)
  metroDistance = getMetroDistance(d)
  floor, floors, isLastFloor, houseType = getHouseTypeAndFloor(d)
  balcony = getBalcony(d)
  wc = getWC(d)
  area = getArea(d)

  #Объединяем в один лист
  out = list(roomsCountType)
  out.append(metro)
  out.extend(metroType)
  out.extend(metroDistance)
  out.extend(floor)
  out.extend(floors)
  out.append(isLastFloor)
  out.extend(houseType)
  out.extend(balcony)
  out.extend(wc)
  out.append(area)

  return out

#Генерируем обучающаюу выборку - xTrain
def getXTrain(data):

  #Получаем строку во всеми вариантами метро
  allMertroNames = list(df["Метро / ЖД станции"].unique())

  #Всевращаем все строки в data1 в векторы параметров и записываем в xTrain
  xTrain = [getAllParameters(d, allMertroNames) for d in data]
  xTrain = np.array(xTrain)

  return xTrain

#Генерируем обучающую выборку - yTrain
def getYTrain(data):

  #Зашружаем лист всех цен квартир по всем строкам data1
  costList = [getCost(d) for d in data]
  yTrain = np.array(costList)

  return yTrain
```

```python
# Инициализация модели с несколькими входными и выходными данными.
model = ak.AutoModel(
    inputs=[ak.TextInput(), ak.StructuredDataInput()],
    outputs=[
        ak.RegressionHead(metrics=["mae"])
    ],
    overwrite=True,
    max_trials=2,
)
# Обучаем модель на подготовленных данных.
model.fit(
    [texts_all_np, xTrain],
    yTrain,
    epochs=3,
)
```
